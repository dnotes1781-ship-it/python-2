# Finding and Removing Outliers

```python
df=pd.read_csv('auto-mpg.csv')
df
```

```python
Q1=df['acceleration'].quantile(0.25)
Q3=df['acceleration'].quantile(0.75)
IQR=Q3-Q1
Low=Q1-(1.5*IQR)
High=Q3+(1.5*IQR)
```

```python
print("Low=",Low)
print("High=",High)
```

```python
Find_Outliers=df[(df['acceleration']<Low) | (df['acceleration']>High)]
Find_Outliers
```

# Removing Outliers

```python
Q1=df['acceleration'].quantile(0.25)
Q3=df['acceleration'].quantile(0.75)
IQR=Q3-Q1
Low=Q1-(1.5*IQR)
High=Q3+(1.5*IQR)
```

```python
Remove_Outliers=df[(df['acceleration']>=Low) & (df['acceleration']<=High)]
Remove_Outliers
```

left_on,right_on

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2,3],'name':['A','B','C'],'dept_id':[10,20,30]})
departments=pd.DataFrame({'d_id':[10,20,40],'dept_name':['HR','IT','Finance']})
pd.merge(employee,departments,left_on='dept_id',right_on='d_id')
```

merge using index
(left_index.right_index)

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2],'name':['A','B'],'dept_id':[10,20]})
departments=pd.DataFrame({'d_id':[10,20],'dept_name':['HR','IT']})
pd.merge(employee,departments,left_index=True,right_index=True)
```

suffixes

```python
import pandas as pd
employee=pd.DataFrame({"emp_id":[1,2],'name':['A','B'],'dept_id':[10,20]})
departments=pd.DataFrame({'dept_id':[10,20],'dept_name':['HR','IT'],'name':['X','Y']})
pd.merge(employee,departments,on='dept_id',suffixes=('_emp','_dept'))
```
