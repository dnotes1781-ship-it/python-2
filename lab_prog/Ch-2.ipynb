Networks

unidirected graph

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(4,5),(5,6),(6,1)])

nx.draw_networkx(G,node_size=1000,node_color='red',width=5,edge_color='blue')
plt.show()
```

Directed Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(4,5),(5,6),(6,1),(2,1)])

nx.draw_networkx(G,node_size=1000,node_color='red',width=2,edge_color='blue')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 
              'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 
              'person5': [1,1,0,0,0]}

Node=list(instagram.keys())
G.add_nodes_from(Node)
Follow=list(instagram.values())
for i in range(len(Follow)):
    for j in range(len(Follow)):
        if Follow[i][j]==1:
            G.add_edge(Node[i],Node[j])
nx.draw_networkx(G,node_size=2200,node_color='red',width=2,edge_color='blue')
plt.show()
```

pie Chart

```python
import pandas as pd
import plotly.express as px
data={'Department':['IT','HR','Finance','Marketing'],
     'Employees':[40,20,25,15]}
df=pd.DataFrame(data)
fig=px.pie(df,values='Employees',names='Department',
          color_discrete_sequence=['blue','green','orange','red'],
          title='Employee Distribution by Department')
fig.update_traces(
    textinfo='percent+label',textposition='inside',pull=[0,0.1,0,0])
fig.show()
```

Histogram

```python
data={'Marks':[40,45,50,55,60,60,65,70,75,80,85],
     'Gender':['Male','Female','Male','Female','Male','Female',
              'Male','Female','Male','Female','Male']}
df=pd.DataFrame(data)
fig=px.histogram(df,x='Marks',color='Gender',
                histfunc='count',pattern_shape='Gender',
                text_auto=True,nbins=10,
                color_discrete_map={'Male':'blue','Female':'pink'})
fig.show()
```
