# parallel coordinates
    - syntax : 
        parallel_coordinates(dataframe ,class_column,col=None,color=None)

```python
import matplotlib.pyplot as plt
from pandas.plotting  import parallel_coordinates
pll = parallel_coordinates(df,'cylinders',cols = ['acceleration' ,'mpg'],color =('red','blue','green','black','orange'))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pll = pd.plotting.parallel_coordinates(df,'cylinders',cols = ['acceleration' ,'mpg'],color =('red','blue','green','black','orange'))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pll = pd.plotting.parallel_coordinates(df,'cylinders',cols = ['acceleration' ,'mpg'],color =('red','blue')) #taking only 2 color , it repeates 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pll = pd.plotting.parallel_coordinates(df,'cylinders',cols = ['acceleration' ,'mpg','displacement'],color =('red','blue','green','black','orange'))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
pll = pd.plotting.parallel_coordinates(df,'origin',cols = ['acceleration' ,'mpg','displacement'],color =('red','blue','green'))
plt.show()
```
